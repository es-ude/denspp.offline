```python
import numpy as np
```

```python
def movmean(arr: np.ndarray, n:int):
    result = []
    for idx, num in enumerate(arr[0]):
        sub = arr[0][max(idx - n, 0): idx + 1]
        print(idx, sub)
        avg = np.mean(sub)
        result.append(avg)
    return np.array([result])
```

```python
a = np.arange(1,11).reshape(1,10)
a
```

```python
movmean(a, 2)
```
